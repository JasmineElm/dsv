# Import libraries

The cell beneath this one is a `code` cell.  It is used to run code.  This cell imports the libraries that we will use in this notebook.  The syntax for importing a library is simply `import` followed by the name of the library.  The `as` keyword is used to give the library an alias.  This is optional but it is common practice to do this as it allows us to refer to the library by a shorter name. this has the added benefit of making our code easier to read and maintain.

```python
import pandas as pd             # Pandas allows us to work with dataframes
import numpy as np              # Numpy is a library for numerical computing
import matplotlib.pyplot as plt # Matplotlib is a plotting library
```

# Create a sample dataset

The next cell creates a sample dataset with 1000 rows and 2 columns. Each cell contains a random number drawn from a standard normal distribution.  There's a lot to unpack in this cell, so let's break it down step by step:

+ `np.random.randn(1000, 2)` generates a 2D array of shape (`(1000, 2)`) filled with random numbers drawn from a standard normal distribution (mean = 0, standard deviation = 1), using numpy's `random.randn function`.
+ `pd.DataFrame(...)` creates a DataFrame from the 2D array generated in the previous step. The `columns=['X', 'Y']` argument names the columns of the DataFrame as 'X' and 'Y'.
+ we assign this dataframe the name `df` using the `df = ` syntax.
+ `df.head()` displays the first 5 rows of the DataFrame. We could also use `df.sample(5)` to display 5 random rows, `df.tail()` to display the last 5 rows or simplu `df` to display the entire DataFrame.

```python
# create a DataFrame using random values from a normal distribution
df = pd.DataFrame(np.random.randn(1000, 2), columns=['X', 'Y'])
# show the first few rows of the DataFrame
df.head()
```

# Visualise the dataset

Visualisation is trivial using `matplotlib`.  The `plt.scatter` function creates a scatter plot of the values in column 'X' on the x-axis and the values in column 'Y' on the y-axis.  The `c='blue` argument tells matplotlib to colour the markers blue. 

The `plt.show()` function displays the plot.

```python
# plot the DataFrame using matplotlibs
plt.scatter(df['X'], df['Y'], c='blue')
plt.show()
```
